```python
import urllib.request as request
import json
import threading
import pandas as pd
import datetime
import time
import math

import matplotlib.pyplot as plt
```

```python
# Метод майнинга информации об автобусе
def run_check():
    df = pd.read_excel('dataset.xlsx')
    df = df[df.columns.drop(list(df.filter(regex='^Unnamed')))]
    url = "http://data.kzn.ru:8082/api/v0/dynamic_datasets/bus/4262.json"
    req = request.Request(url)
    r = request.urlopen(req).read()
    data = json.loads(r.decode('utf-8'))
    length = len(df)
    df.loc[length, 'GaragNumb'] = data['data']['GaragNumb']
    df.loc[length, 'TimeNav'] = data['data']['TimeNav']
    df.loc[length, 'Latitude'] = data['data']['Latitude']
    df.loc[length, 'Longitude'] = data['data']['Longitude']
    df.loc[length, 'Azimuth'] = data['data']['Azimuth']
    df.loc[length, 'Speed'] = data['data']['Speed']
    df.loc[length, 'Marsh'] = data['data']['Marsh']
    df.loc[length, 'Graph'] = data['data']['Graph']
    df.to_excel('dataset.xlsx')
```

```python
# здесь вызывается метод и майнит нужные мне данные, которые я потом буду обрабатывать
morning = datetime.datetime(2020, 10, 3, 6, tzinfo=None)
evening = datetime.datetime(2020, 10, 4, 1, tzinfo=None)
now = datetime.datetime.now()
while (morning < now and evening > now):
    run_check()
    now = datetime.datetime.now()
    print('отработал {}'.format(now))
    time.sleep(60)
```

Сначала посчитаем нужные нам значения для 27 числа.

```python
for i in range(len(df)):
#     if in this day
    first_date = datetime.datetime(2020, 9, 27, 1, tzinfo=None)
    second_date = datetime.datetime(2020, 9, 27, 23, tzinfo=None)
    if ((df['TimeNav'][i] < second_date) and (df['TimeNav'][i] > first_date)):
#         if in the bus-stop area
        if ((df['Latitude'][i] > 55.733) & (df['Latitude'][i] < 55.734) & (df['Longitude'][i] > 49.135) & (df['Longitude'][i] < 49.136)):
            df['distance'][i] = 0
            df['time'][i] = 0
#        if not in the bus-stop area
        else:
#         рассчитываем расстояние автобуса
            x2 = df['Longitude'][i]
            y2 = df['Latitude'][i]
            x1 = df['Longitude'][i - 1]
            y1 = df['Latitude'][i - 1]
            current_distance = math.sqrt(math.pow((x2 - x1), 2) + math.pow((y2 - y1), 2))
            df['distance'][i] = df['distance'][i - 1] + current_distance
#             рассчитываем затраченное время
            current_time = df['TimeNav'][i] - df['TimeNav'][i - 1]
            current_time = current_time.seconds / 60
            df['time'][i] = df['time'][i - 1] + current_time
```

Метод, показывающий через сколько времени автобус приедет на остановку

```python
def get_bus_to_area_time():
#     считываем время сейчас и перерабатываем его в часы/минуты
    time_now = datetime.datetime.now()
    time_now = datetime.time(hour = time_now.hour, minute = time_now.minute)
#     смотрим ближайшие в датасете к нашему времени
    time_set = new_df[new_df['simple_time'] > time_now]
    time_set = time_set.reset_index(drop=True)
#     если такое время есть в датасете, то делаем дальше
    if (len(time_set) > 0):
        indexes = time_set.index.tolist()
        current_index = indexes[0]
#         вычисляем через сколько времени приедет автобус
        for i in range(current_index + 1, len(new_df)):
            if ((new_df['time'][i] == 0) and (new_df['time'][i - 1] != 0)):
                return new_df['time'][i - 1] - time_set['time'][0]
    return 'невозможно определить, время не соответствует датасету'
```
