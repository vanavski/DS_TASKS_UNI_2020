```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
# load iris dataset
# iris = datasets.load_iris()
# convert to dataframe (Just for easy access) 
iris_dataset = pd.read_excel('df.xlsx')
# iris_dataset = pd.DataFrame(iris.data)
# load iris labels 
iris_labels = iris_dataset.exam

# iris_dataset[iris_dataset['gender'] == 'М'] = 0
# iris_dataset[iris_dataset['gender'] == 'Ж'] = 1

for i in range(len(iris_dataset)):
    if(iris_dataset['gender'][i] == 'М'):
        iris_dataset.loc[i, 'gender'] = 1
    else: iris_dataset.loc[i, 'gender'] = 0
        
for i in range(len(iris_dataset)):
    if(iris_dataset['exam'][i] == 'Да'):
        iris_dataset.loc[i, 'exam'] = 1
    else: iris_dataset.loc[i, 'exam'] = 0

iris_dataset.head()

```

```python
# less use standary scaler on the data
from sklearn.preprocessing import StandardScaler

standard = StandardScaler()
iris_dataset_standar = pd.DataFrame(standard.fit_transform(iris_dataset))
iris_dataset_standar.head()
```

```python
from som import SOM

# initialize the SOM
som = SOM(15, 15)

# fit the SOM for 20000 epochs, save the error every 300 steps
som.fit(iris_dataset_standar.to_numpy(), 300, save_e=True, interval=50) 
```

```python
# plot the training error history
som.plot_error_history()
```
